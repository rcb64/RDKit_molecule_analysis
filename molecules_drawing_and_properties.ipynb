```python
from rdkit import Chem
from rdkit.Chem import AllChem

# Create a molecule from SMILES
mol = Chem.MolFromSmiles('O=C(C[C@@H](N(C(C(C=C1)=CC(Cl)=C1C2=C(OCC3=CC=CC=C3)C=CC(C(F)(F)F)=C2)=O)[H])C(NCC4=CN(CC(OC5=CC=C(F)C=C5)=O)N=N4)=O)OC(C)(C)C')

# Add 3D coordinates to the molecule
AllChem.EmbedMolecule(mol)

# Access the conformer and check atomic positions
conformer = mol.GetConformer()
for i in range(mol.GetNumAtoms()):
    pos = conformer.GetAtomPosition(i)
    print(f"Atom {i}: x={pos.x}, y={pos.y}, z={pos.z}")
```

```python
from rdkit.Chem import DataStructs
fp = Chem.RDKFingerprint(mol)
```

```python
smiles_list = ["CCO", "CCCC", "CCN", "C1=CC=CC=C1", "O=C(C[C@@H](N(C(C(C=C1)=CC(Cl)=C1C2=C(OCC3=CC=CC=C3)C=CC(C(F)(F)F)=C2)=O)[H])C(NCC4=CN(CC(OC5=CC=C(F)C=C5)=O)N=N4)=O)OC(C)(C)C"]  # Ethanol, Butane, Ethylamine, Benzene
names = ["Ethanol", "Butane", "Ethylamine", "Benzene", "Radu's molecule"]
from rdkit import Chem
from rdkit.Chem import Descriptors, Draw

for smile, name in zip(smiles_list, names):
    mol = Chem.MolFromSmiles(smile)
    mw = Descriptors.MolWt(mol)
    h_donors = Descriptors.NumHDonors(mol)
    print(f"{name}: Molecular Weight = {mw}, H-bond Donors = {h_donors}")
```

```python
from rdkit.Chem import Draw
mols = [Chem.MolFromSmiles(smile) for smile in smiles_list]
Draw.MolsToGridImage(mols, legends=names)
```

```python
import pandas as pd

data = []
for smile, name in zip(smiles_list, names):
    mol = Chem.MolFromSmiles(smile)
    mw = Descriptors.MolWt(mol)
    h_donors = Descriptors.NumHDonors(mol)
    data.append({"Name": name, "SMILES": smile, "Molecular Weight": mw, "H-bond Donors": h_donors})

df = pd.DataFrame(data)
df.to_csv("molecule_properties.csv", index=False)
```

```python
from rdkit.Chem import DataStructs
mol1 = Chem.MolFromSmiles("O=C(C[C@@H](N(C(C(C=C1)=CC(Cl)=C1C2=C(OCC3=CC=CC=C3)C=CC(C(F)(F)F)=C2)=O)[H])C(NCC4=CN(CC(OC5=CC=C(F)C=C5)=O)N=N4)=O)OC(C)(C)C")
mol2 = Chem.MolFromSmiles("O=C(C[C@@H](N(C(C(C=C1)=CC(Cl)=C1C2=C(O)C=CC(C(F)(F)F)=C2)=O)[H])C(NCC3=CN(CC(OC4=CC=C(F)C=C4)=O)N=N3)=O)O")
fp1 = Chem.RDKFingerprint(mol1)
fp2 = Chem.RDKFingerprint(mol2)
similarity = DataStructs.FingerprintSimilarity(fp1, fp2)
print(f"Similarity: {similarity}")
```

```python
import matplotlib.pyplot as plt
molecular_weights = [Descriptors.MolWt(Chem.MolFromSmiles(smile)) for smile in smiles_list]
plt.bar(names, molecular_weights)
plt.ylabel("Molecular Weight")
plt.show()
```
